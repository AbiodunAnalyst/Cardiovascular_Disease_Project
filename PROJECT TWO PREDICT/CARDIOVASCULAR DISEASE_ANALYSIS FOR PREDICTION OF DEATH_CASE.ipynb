## Mann-Whitney test 

```python
import scipy.stats as stats
```

**Null Hypothesis (H0):** There is no difference between the distributions of the two groups.

**Alternative Hypothesis (H1):** There is a significant difference between the distributions of the two groups.

**Decision Rule:**
    if p_value < alpha
    Reject H0 and conclude that there is a significant difference between the distributions of the two groups

```python
# Run Mann-Whitney U test
statistic, p_value = stats.mannwhitneyu(death_df,alive_df, alternative='two-sided')


# Print the results
print(f'Mann-Whitney U statistic: {statistic}')
print(f'P-value: {p_value}')


# Interpret the results
alpha = 0.05
if any(p_value < alpha):
    print('Reject the null hypothesis. There is a significant difference between the distribution of the two groups.')
else:
    print('Fail to reject the null hypothesis. There is no significant difference between the distribution of the two groups.')
```

## Chi-Square test

```python
from scipy.stats import chisquare
```

**H0 :** There is no association between the two categorical variables.

**H1 :** There is an association between the two categorical variables.
    
**Decision Rule:** if p_value < alpha Reject H0

```python
# Count the frequencies of each category
observed_counts = df_data['DEATH_EVENT'].value_counts()

# If you have an expected distribution (e.g., equal distribution)
expected_counts = len(df_data) / len(observed_counts)

# Perform Chi-Square test
chi2_stat, p_value = chisquare(observed_counts, f_exp=expected_counts)

# Print the results
print(f'Chi-Square Statistic: {chi2_stat}')
print(f'P-value: {p_value}')

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print('Reject the null hypothesis. There is a significant association between the two category.')
else:
    print('Fail to reject the null hypothesis. There is no significant association between the two category.')
```

### Separate dataset into train and test

```python
# separate dataset into train and test

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 0)
X_train.shape, X_test.shape
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# set up Tomek Links

tl = TomekLinks(
    sampling_strategy='auto',  # undersamples only the majority class
    n_jobs=4)  

X_resampled_t, y_resampled_t = tl.fit_resample(X_train, y_train)
```

```python
X_resampled_t.shape, y_resampled_t.shape
```

```python
# set up Border line SMOTE

sm_bl = BorderlineSMOTE(
    sampling_strategy='auto',  # samples only the minority class
    random_state=0,  # for reproducibility
    k_neighbors=5, # the neighbours to crete the new examples
    m_neighbors=10, # the neiighbours to find the DANGER group
    kind='borderline-1',
    n_jobs=4
)

X_resampled_b, y_resampled_b = sm_bl.fit_resample(X_train, y_train)
```

```python
X_resampled_b.shape, y_resampled_b.shape
```

### RANDOM FOREST MODEL

```python
def run_randomForests(X_train, X_test, y_train, y_test):
    
    rf = RandomForestClassifier(random_state = 42)
    rf.fit(X_train, y_train)
    
    
    pred = rf.predict(X_test)
    
    
    sns.reset_orig()

    ConfusionMatrixDisplay.from_predictions(
            y_test, pred,
            labels = rf.classes_,
            cmap = 'magma'
    );
    
   
    print('Test')
    print(classification_report(y_test, pred, target_names = ['Alive', 'Dead']))
    
    
    print('Train set')
    pred = rf.predict_proba(X_train)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_train, pred[:,1])))
    
    print('Test set')
    pred = rf.predict_proba(X_test)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_test, pred[:,1])))
```

### Evaluate performance of algorithm built using imbalanced dataset

```python
# evaluate performance of algorithm built using imbalanced dataset

run_randomForests(X_train,
                  X_test,
                  y_train,
                  y_test)
```

### IMPROVING THE MODEL USING Gridsearch FOR RANDOM FOREST MODEL

```python
def grid_randomForests(X_train, X_test, y_train, y_test):
    
    rf_classifier = RandomForestClassifier(random_state = 42)
    
    grid_search = GridSearchCV(rf_classifier, param_grid, cv=5)
    #grid_search.best_params_
    grid_search.fit(X_train, y_train)
    
    
    grid_predictions = grid_search.predict(X_test)
    
    sns.reset_orig()

    ConfusionMatrixDisplay.from_predictions(
            y_test, grid_predictions,
            labels = grid_search.classes_,
            cmap = 'magma'
    );
    
    
    
    print(classification_report(y_test, grid_predictions, target_names = ['Alive', 'Dead']))
    
    
    print('Train set')
    grid_predictions = grid_search.predict_proba(X_train)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_train, grid_predictions[:,1])))
    
    print('Test set')
    grid_predictions = grid_search.predict_proba(X_test)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_test, grid_predictions[:,1])))

```

### IMPROVING THE MODEL USING Gridsearch FOR VECTOR SUPPORT MODEL

```python
# Define the parameter grid
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [0.0001, 0.001, 0.01, 0.1, 1], 'kernel': ['linear', 'rbf']}
```

```python
def grid_SupportVector(X_train, X_test, y_train, y_test):
    
    svm_classifier = SVC(probability=True,  random_state=42)
    
    grid_search = GridSearchCV(svm_classifier, param_grid, cv=5)
    
    grid_search.fit(X_train, y_train)
    
    
    grid_predictions = grid_search.predict(X_test)
    
    sns.reset_orig()

    ConfusionMatrixDisplay.from_predictions(
            y_test, grid_predictions,
            labels = grid_search.classes_,
            cmap = 'magma'
    );
    
    
    
    print(classification_report(y_test, grid_predictions, target_names = ['Alive', 'Dead']))
    
    
    
    print('Train set')
    grid_predictions = grid_search.predict_proba(X_train)
    print('Suport vector roc-auc: {}'.format(roc_auc_score(y_train, grid_predictions[:,1])))
    
    print('Test set')
    grid_predictions = grid_search.predict_proba(X_test)
    print('Suport vector roc-auc: {}'.format(roc_auc_score(y_test, grid_predictions[:,1])))

```

```python
class visualization:
    def __init__(self, df_data):
        self.df_data = df_data
    
    def missing_value(self): # check for missing value
        sns.heatmap(self.df_data.isnull(), yticklabels = False, cbar =False, cmap = 'Blues')
        
    def distribution(self): # checking the distribution of the variable
        self.df_data.hist(bins = 30, figsize = (20,20), color = 'b')
        
    def age_distribution_death_case(self): # checking age distribution
        plt.figure(figsize = [25,12])
        sns.countplot(x = 'age', hue = 'high_blood_pressure', data = self.df_data)
        
    def gender_anaemia(self):# checking gender on HBP
        plt.figure(figsize = [15,30])
        plt.subplot(411)
        sns.countplot(x = 'gender', hue = 'anaemia', data = self.df_data)

    def HBP_plot_anaemia(self):# checking if there is effect of anaemia on HBP
        plt.figure(figsize = [15,30])
        plt.subplot(412)
        sns.countplot(x = 'high_blood_pressure', hue = 'anaemia', data = self.df_data)

    def diabetes_anaemia_plot(self):# checking if there is effect of diabetes on HBP
        plt.figure(figsize = [15,30])
        plt.subplot(413)
        sns.countplot(x = 'diabetes', hue = 'anaemia', data = self.df_data)

    def smoking_anaemia_plot(self):# checking if there is effect of smoking on HBP
        plt.figure(figsize = [15,30])
        plt.subplot(414)
        sns.countplot(x = 'smoking', hue = 'anaemia', data = self.df_data)
   
    def distribution_plot_anaemia(self):# checking possible pattern for predicting HBP
        sns.pairplot(df_data, hue = 'anaemia')


visual_plot = visualization(df_data)
```

```python
class KDEPlotter:
    def __init__(self, anaemia_df, no_anaemia_df):
        self.anaemia_df = anaemia_df
        self.no_anaemia_df = no_anaemia_df
    
    def kdeplot_time(self):
        plt.figure(figsize = (10,7))
        sns.kdeplot(self.anaemia_df['time'], label = 'Population that have HBP', shade = True, color ='r')
        sns.kdeplot(self.no_anaemia_df['time'], label = "Population that don't have anaemia", shade = True, color ='b')

        plt.xlabel('Heart Failure Time Taken ')
        plt.legend()
        plt.show()

    def kdeplot_creatinine_phosphokinase(self):
        plt.figure(figsize = (10,7))
        sns.kdeplot(anaemia_df['creatinine_phosphokinase'], label = 'Population that have anaemia', shade = True, color ='r')
        sns.kdeplot(no_anaemia_df['creatinine_phosphokinase'], label = "Population that don't have anaemia", shade = True, color ='b')

        plt.xlabel('creatinine phosphokinase rate ')
        plt.legend()
        plt.show()

    def kdeplot_ejection_fraction(self):
        plt.figure(figsize = (10,7))
        sns.kdeplot(anaemia_df['ejection_fraction'], label = "Population that have anaemia", shade = True, color ='r')
        sns.kdeplot(no_anaemia_df['ejection_fraction'], label = "Population that don't have anaemia", shade = True, color ='b')

        plt.xlabel('ejection fraction rate ')
        plt.legend()
        plt.show()

    def kdeplot_platelets(self):
        plt.figure(figsize = (10,7))
        sns.kdeplot(anaemia_df['platelets'], label = "Population that have anaemia", shade = True, color ='r')
        sns.kdeplot(no_anaemia_df['platelets'], label = "Population that don't have anaemia", shade = True, color ='b')

        plt.xlabel('platelets rate ')
        plt.legend()
        plt.show()


    def kdeplot_serum_creatinine(self):    
        plt.figure(figsize = (10,7))
        sns.kdeplot(anaemia_df['serum_creatinine'], label = "Population that have anaemia", shade = True, color ='r')
        sns.kdeplot(no_anaemia_df['serum_creatinine'], label = "Population that don't have anaemia", shade = True, color ='b')

        plt.xlabel('serum creatinine rate ')
        plt.legend()
        plt.show()


    def kdeplot_serum_sodium(self):    
        plt.figure(figsize = (10,7))
        sns.kdeplot(anaemia_df['serum_sodium'], label = "Population that have anaemia", shade = True, color ='r')
        sns.kdeplot(no_anaemia_df['serum_sodium'], label = "Population that don't have anaemia", shade = True, color ='b')

        plt.xlabel('serum_sodium rate ')
        plt.legend()
        plt.show()

kde_plotter = KDEPlotter(anaemia_df, no_anaemia_df)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler

import warnings
warnings.filterwarnings("ignore")
```

```python
df_data.columns
```

```python
df_data_ann = df_data[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes',
       'ejection_fraction', 'high_blood_pressure', 'platelets',
       'serum_creatinine', 'serum_sodium', 'gender', 'smoking', 'time']]
```

```python
df_data_ann.head(2)
```

```python
y = df_data_ann['anaemia']
X = df_data_ann.drop(labels=['anaemia'], axis=1)
```

### Separate dataset into train and test

```python
# separate dataset into train and test

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 0)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Data Normalisation and Data scaling
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# set up Tomek Links

tl = TomekLinks(
    sampling_strategy='auto',  # undersamples only the majority class
    n_jobs=4)  

X_resampled_t, y_resampled_t = tl.fit_resample(X_train, y_train)
```

```python
X_resampled_t.shape, y_resampled_t.shape
```

```python
# set up Border line SMOTE

sm_bl = BorderlineSMOTE(
    sampling_strategy='auto',  # samples only the minority class
    random_state=0,  # for reproducibility
    k_neighbors=5, # the neighbours to crete the new examples
    m_neighbors=10, # the neiighbours to find the DANGER group
    kind='borderline-1',
    n_jobs=4
)

X_resampled_b, y_resampled_b = sm_bl.fit_resample(X_train, y_train)
```

```python
X_resampled_b.shape, y_resampled_b.shape
```

```python
# set up SMOTE

sm = SMOTE(
    sampling_strategy='auto',  # samples only the minority class
    random_state=0,  # for reproducibility
    k_neighbors=5,
    n_jobs=4
)

X_resampled_s, y_resampled_s = sm.fit_resample(X_train, y_train)
```

```python
X_resampled_b.shape, y_resampled_b.shape
```

```python
# set up Random oversampler

ros = RandomOverSampler(
    sampling_strategy='not majority', # samples all but majority class
    random_state=0,  # for reproducibility
)  

X_resampled_ros, y_resampled_ros = ros.fit_resample(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
from yellowbrick.classifier import (
    ClassificationReport,
    DiscriminationThreshold,
)
```

### RANDOM FOREST MODEL

```python
def run_randomForests(X_train, X_test, y_train, y_test):
    
    # Create the RandomForestClassifier model
    rf = RandomForestClassifier(random_state = 42)
    rf.fit(X_train, y_train)
    
    
    pred = rf.predict(X_test)
    
    
    sns.reset_orig()

    ConfusionMatrixDisplay.from_predictions(
            y_test, pred,
            labels = rf.classes_,
            cmap = 'magma'
    );
    
   
    print('Test')
    print(classification_report(y_test, pred, target_names = ['No Anaemia', 'Anaemia']))
    
    
    print('Train set')
    pred = rf.predict_proba(X_train)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_train, pred[:,1])))
    
    print('Test set')
    pred = rf.predict_proba(X_test)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_test, pred[:,1])))
```

### Evaluate performance of algorithm built using imbalanced dataset

```python
# evaluate performance of algorithm built using imbalanced dataset

run_randomForests(X_train,
                  X_test,
                  y_train,
                  y_test)
```

```python
# evaluate performance of algorithm built using Tomek Links

run_randomForests(X_resampled_t,
                  X_test,
                  y_resampled_t,
                  y_test)
```

### IMPROVING THE MODEL USING Gridsearch FOR RANDOM FOREST MODEL

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [10, 20, 50,100, 200],
    'max_depth': [None, 1, 2, 3, 4],
    'min_samples_split': [0.1, 0.3, 0.5, 1, 3 , 5]
}
```

```python
def grid_randomForests(X_train, X_test, y_train, y_test):
    
    rf_classifier = RandomForestClassifier(random_state = 42)
    
    grid_search = GridSearchCV(rf_classifier, param_grid, cv=5)
    #grid_search.best_params_
    grid_search.fit(X_train, y_train)
    
    
    grid_predictions = grid_search.predict(X_test)
    
    sns.reset_orig()

    ConfusionMatrixDisplay.from_predictions(
            y_test, grid_predictions,
            labels = grid_search.classes_,
            cmap = 'magma'
    );
    
    
    
    print(classification_report(y_test, grid_predictions, target_names = ['No Anaemia', 'Anaemia']))
    
    
    print('Train set')
    grid_predictions = grid_search.predict_proba(X_train)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_train, grid_predictions[:,1])))
    
    print('Test set')
    grid_predictions = grid_search.predict_proba(X_test)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_test, grid_predictions[:,1])))

```
